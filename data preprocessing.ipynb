```python
import pandas as pd
import numpy as np
df=pd.read_csv("/Users/irene/Downloads/Bankruptcy1.csv")
df
```

```python
df
```

```python
X=df.drop(['Bankrupt?', 'Unnamed: 0'], axis=1)
for i in X.columns:
    X[i]=X[i].fillna(X[i].mean())
```

```python
for i in X.columns:
    print(X[i].isnull().sum())
```

```python
y=df['Bankrupt?']
print(y.shape[0], y[y==0].shape[0], y[y==1].shape[0])
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42)
X_res, y_res=sm.fit_resample(X, y)
print('SMOTE Resampled dataset shape %s' % Counter(y_res))
```

```python
#feature selection
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
X_=pca.fit_transform(X_res, y_res)
X=pd.DataFrame(X_)
X_.shape[1]
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectFromModel
rf=RandomForestClassifier()
rf.fit(X_res, y_res)
importance=rf.feature_importances_
for i, v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i, v))
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

```python
#排序
plt.figure(figsize=(50, 50))
importances=pd.DataFrame({'feature':X_res.columns, 'importance':np.round(rf.feature_importances_,3)})
importances=importances.sort_values('importance', ascending=False).set_index('feature')
importances.plot.bar()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
silhouette_avgs=[]
ks=range(2, 11)
for k in ks:
    kmeans_fit=KMeans(n_clusters=k).fit(X)
    cluster_labels=kmeans_fit.labels_
    silhouette_avg=metrics.silhouette_score(X, cluster_labels)
    silhouette_avgs.append(silhouette_avg)
plt.bar(ks, silhouette_avgs)
plt.show()
```
